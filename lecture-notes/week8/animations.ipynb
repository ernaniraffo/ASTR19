# Simple Animations for ASTR 19
Let's build from this to make some animations.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from tqdm.autonotebook import tqdm
%matplotlib inline
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

```python
fig = plt.figure(linewidth=1, figsize=(10, 5))
ax = plt.subplot(1, 1, 1, aspect=1)

ax.plot(X, C, zorder=9, color='#1f77b4', label='cos(x)')
ax.plot(X, C, 'o', markevery=10, zorder=10, color='#1f77b4', ms=6)
ax.plot(X, C, 'o', markevery=10, zorder=11, color='white', ms=3)

ax.plot(X, S, zorder=9, color='#ff7f0e', label='sin(x)')
ax.plot(X, S, 'o', markevery=10, zorder=10, color='#ff7f0e', ms=6)
ax.plot(X, S, 'o', markevery=10, zorder=11, color='white', ms=3)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='upper right', frameon=True)

plt.show()
```

Now let's try animating a curve and save as a gif

```python
# Create figure as usual
from code import interact


fig = plt.figure(linewidth=1, figsize=(10, 5))
ax = plt.subplot(1, 1, 1, aspect=1)

# Extract lines and points as objects of a class in matplotlib
(line1, ) = ax.plot(X, C, zorder=9, color='#1f77b4', label='cos(x)')
(dots1, ) = ax.plot(X, C, 'o', markevery=10, zorder=10, color='#1f77b4', ms=6)
(wdots, ) = ax.plot(X, C, 'o', markevery=10, zorder=11, color='white', ms=3)

(line2, ) = ax.plot(X, S, zorder=9, color='#ff7f0e', label='sin(x)')
(dots2, ) = ax.plot(X, S, 'o', markevery=10, zorder=10, color='#ff7f0e', ms=6)
(wdots2, ) = ax.plot(X, S, 'o', markevery=10, zorder=11, color='white', ms=3)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='upper right', frameon=True)

# Function that updates the plot based on the frame number
def update(frame):
    # data used for line1 is now X and C
    # from 0th to (frame-1)th element
    line1.set_data(X[:frame], C[:frame])
    dots1.set_data(X[:frame], C[:frame])
    wdots.set_data(X[:frame], C[:frame])

    # data used for line2 is now X and S
    # from 0th to (frame-1)th element
    line2.set_data(X[:frame], S[:frame])
    dots2.set_data(X[:frame], S[:frame])
    wdots2.set_data(X[:frame], S[:frame])

# remove whitespace
plt.tight_layout()

# to determine what the output animation file is
# choose the writer function animation
# this selects GIF
writer = animation.ImageMagickWriter(fps=30)
ani = animation.FuncAnimation(fig, update, interval=10, frames=len(X))

# create a progress bar
bar = tqdm(total=len(X))

# create the GIF
ani.save('sin-cos-animated.gif', writer=writer, dpi=100, progress_callback=lambda i, n: bar.update(1))

bar.close()
```
