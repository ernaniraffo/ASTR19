# Make a figure with geometric shapes

```python
import matplotlib.pyplot as plt
from matplotlib import patches
%matplotlib inline
```

```python
# create a figure
f = plt.figure(figsize = (6,6))

# add a single subplot to the figure
ax = f.add_subplot(111)

# remove x and y tick labels

# create a rectangle
rectangle = patches.Rectangle((0.05, 0.05), 0.4, 0.4, edgecolor='red', facecolor='black', alpha=0.9)

ax.add_patch(rectangle)

# create a circle
circle = patches.Circle((0.6, 0.6), 0.3, edgecolor='red', facecolor='black', alpha=0.9)

ax.add_patch(circle)

# create an ellipse

ellipse = patches.Ellipse((0.9, 0.3), 0.2, 0.4, edgecolor='red', facecolor='black', alpha=0.9)

ax.add_patch(ellipse)

plt.show()
```

```python
# check if a number is odd
def is_odd(num: int | float) -> bool:
    return num % 2 != 0
```

```python
from random import choice

colors = [(choice(range(255))/255, choice(range(255))/255, choice(range(255))/255) for _ in range(2)]
# create a new figure
f = plt.figure(figsize=(6, 6))
ax = f.add_subplot(111)

# iterate over columns and rows to create black rectangles
ROWS = 8
COLUMNS = 8
for i in range(ROWS):
    for j in range(COLUMNS):
        # colors = [(choice(range(255))/255, choice(range(255))/255, choice(range(255))/255) for _ in range(2)]
        if is_odd(i + j + 1):
            rectangle = patches.Rectangle((i, j), 1, 1, edgecolor='black', facecolor='silver', alpha=1)
            ax.add_patch(rectangle)
        else:
            rectangle = patches.Rectangle((i, j), 1, 1, edgecolor='black', facecolor='gold', alpha=1)
            ax.add_patch(rectangle)

# remove x and y ticklables
ax.set_xticks([])
ax.set_yticks([])

ax.set_xlim(0, 8)
ax.set_ylim(0, 8)

plt.show()
```
