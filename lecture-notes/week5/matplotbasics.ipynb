# Plotting basics

First, we import numpy and matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# tell jupyter notebook to display plots
%matplotlib inline
```

Then, define the data we want to plot

```python
n = 100
x = np.linspace(0, 2 * np.pi, n)
y = np.sin(x)
```

Make the plot

```python
# create the elements of the plot
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')

plt.savefig('sinx.png', bbox_inches='tight', dpi=1000)
plt.rcParams.update({
    "text.usetex": True,
})

# show the plot
plt.show()
```
