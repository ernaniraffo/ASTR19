# Plot with polar coordinates

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# when angles are measured in radians, 1 turn is 2 radians

n = 100

theta = np.linspace(0, 2 * np.pi, 50)
y1 = np.ones(50)
y2 = theta * 0.25

# plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='polar')

ax.plot(theta, y1, label='constant')
ax.plot(theta, y2, label='linear')

# add grid
ax.grid(True)

ax.legend(loc=1)

plt.show()
```

```python
def dipole_field(theta):
    return np.sin(theta)**2
```

```python
n = 100
# define theta values
theta = np.linspace(0, 2 * np.pi, n)

plt.figure(figsize=(8, 6))
ax = plt.subplot(111, projection='polar')
ax.plot(theta, dipole_field(theta), label='dipolar field')

ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_rlabel_position(90)

line_length = 0.3
ax.plot([0, np.pi], [line_length, line_length], color='black', linestyle='-', linewidth=5)

ax.grid(True)
ax.legend(loc='upper right')

plt.show()
```
