# ASTR19 Pretty Picture Notebook

The goal of this notebook is to make a pretty picture from some simulate JWST astronomical data. We will create an RGB image and save it as a .png file.

```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
%matplotlib inline
```

```python
# define file names
fdata_F277W = 'session_18/F277W_cutout.fits'
fdata_F356W = 'session_18/F356W_cutout.fits'
fdata_F444W = 'session_18/F444W_cutout.fits'

# read in data
hdu_2 = fits.open(fdata_F277W)
hdu_3 = fits.open(fdata_F356W)
hdu_4 = fits.open(fdata_F444W)

# get the image data
data_2 = hdu_2[0].data
data3 = hdu_3[0].data
data4 = hdu_4[0].data
```

# Have a look at the data

The data arrays are images, so let's have a look at the data.

```python
# First, plot the data as is
f = plt.figure(figsize=(7,7))
plt.imshow(data_2)
plt.colorbar()
plt.show()
```

```python
def rescale_image(data):
    pdata_tmp = data.copy()

    m = np.nanmean(pdata_tmp)

    vpmin = 0.5 * m
    vpmax = 100 * m
    pdata_tmp[pdata_tmp < vpmin] = vpmin
    pdata_tmp[pdata_tmp > vpmax] = vpmax

    pdata_tmp = np.log10(pdata_tmp)
    return pdata_tmp
```

```python
def fits_quicklook(data, fnx=10, fny=10):
    f = plt.figure(figsize=(fnx, fny))
    pdata_tmp = rescale_image(data)
    plt.imshow(pdata_tmp)
```

```python
fits_quicklook(data_2)
```
