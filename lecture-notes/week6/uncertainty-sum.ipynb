# Uncertainty of the sum of two variables

```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
%matplotlib inline
```

```python
# fix the random seed
np.random.seed(42)

n = 50
x = np.random.uniform(-0.5, 0.5, n) # x = 0 +- 0.5
y = np.random.uniform(-0.5, 0.5, n) # y = 0 +- 0.5
z = x + y

# plot histogram of data in 3 different panels
width = 0.1
bins = np.arange(-1.5, 1.5, width)

fig = plt.figure(figsize=(5, 5))
axx = fig.add_subplot(3, 1, 1) # subplot for x histogram
axy = fig.add_subplot(3, 1, 2) # subplot for y histogram
axz = fig.add_subplot(3, 1, 3)

axx.hist(x, bins=bins, alpha=0.5, edgecolor='black')
axx.set_xticks([])
axx.set_ylabel('x frequency')

axy.hist(y, bins=bins, alpha=0.5, edgecolor='black')
axy.set_xticks([])
axy.set_ylabel('y frequency')

axz.hist(z, bins=bins, alpha=0.5, edgecolor='black')
axz.set_ylabel('z frequency')

plt.subplots_adjust(hspace=0.0)

plt.show()

# statistical analysis

rows = ['x', 'y', 'z']
columns = ['mean', 'stdev', 'var']

data = [
    [np.mean(x), np.std(x), np.var(x)],
    [np.mean(y), np.std(y), np.var(y)],
    [np.mean(z), np.std(z), np.var(z)]
]

table = tabulate(data, headers=columns, showindex=rows, tablefmt='fancy_grid')
print(table)
```
