# Exponential and Guassian

```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
%matplotlib inline
```

```python
# create array from 0 to 10
x = np.arange(0, 11, 1)

# create arrays for 2^x, e^x, 10^x
y1 = 2**x
y2 = np.exp(x)
y3 = 10**x

# create a table
data = list(zip(x, y1, y2, y3))
columns = ['x', '2^x', 'e^x', '10^x']

# print the table
table = tabulate(data, headers=columns, tablefmt='fancy_grid')
print(table)
```

```python
#  plot the 3 functions
n = 1000
x = np.linspace(-3, 3, n)

y1 = 2**x
y2 = np.exp(x)
y3 = 10**x

plt.plot(x, y1, label='2^x')
plt.plot(x, y2, label='e^x')
plt.plot(x, y3, label='10^x')
plt.ylim(0, 10)
plt.legend()

plt.show()
```

```python
x = np.linspace(-3, 3, 1000)
mean = 1
sigma = 2
y = np.exp(-(x - mean)**2 / (2 * sigma**2)) * (1 / np.sqrt(2 * np.pi * sigma**2))
plt.plot(x, y)
plt.show()
```

```python
def guassian(x, mean, sigma):
    return np.exp(-(x - mean)**2 / (2 * sigma**2)) * (1 / np.sqrt(2 * np.pi * sigma**2))
```

```python
# create 3 guassian functions
x = np.linspace(-10, 10, 1000)
y1 = guassian(x, 0, 1)
y2 = guassian(x, 1, 2)
y3 = guassian(x, 2, 3)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

How does the guassian function compares to the normal distribution we made earlier?

```python
# generate random data
n = 50000
mean = 0.0
sigma = 0.25
samples = np.random.normal(mean, sigma, n)

# create a histogram
width = 0.05
bins = np.arange(-1.5, 1.5, width)
x = np.linspace(-1.5, 1.5, 1000)

# probability density function
pdf = guassian(x, mean, sigma)

plt.hist(samples, bins=bins, alpha=0.5, edgecolor='black', label='histogram')
plt.plot(x, pdf * n * width, label='density function')
plt.xlabel('x')
plt.ylabel('x frequency')
plt.legend()

plt.show()
```
