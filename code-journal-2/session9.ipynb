```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# use numpy to pull 1000 random numbers distributed uniformly between [0, 1]
nums = np.random.normal(scale=1, size=1000)
print(nums)
```

```python
# histogram the random numbers into 100 bins and plot the histogram
f, ax = plt.subplots(1, 1, figsize=(4, 4))
ax.hist(nums, facecolor='red', edgecolor='black', bins=100, alpha=0.75)

# label our axes
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)

# save the figure
plt.savefig('rand-histogram.pdf', bbox_inches='tight', dpi=1000)
```
