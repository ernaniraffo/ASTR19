```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create an array x ranging from [0, 1] in 100 steps
x = np.linspace(0, 1, 100)
```

```python
# define functions that return sin(x) and cos(x)
def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)
```

```python
# create a multipanel plot (1 row, 2 columns)
nrow = 1
ncolumn = 2
f, ax = plt.subplots(1, 2)

# plot sin(x) vs x in the left panel
ax[0].plot(x, sin(x))
ax[0].set_title('sin(x)')

# plot cos(x) vs x in the right panel
ax[1].plot(x, cos(x))
ax[1].set_title('cos(x)')

# fix layout
plt.tight_layout()

# save the figure
plt.savefig('sin-cos.pdf', bbox_inches='tight', dpi=1000)
```
