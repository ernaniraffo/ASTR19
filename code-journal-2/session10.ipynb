```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate x axis
x = np.random.rayleigh(size=1000)
```

```python
# histogram the random numbers into 100 bins
f, ax = plt.subplots(1, 1, figsize=(4, 4))
ax.hist(x, bins=100, facecolor='red', edgecolor='black')

# label our axes
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)

# save the figure
plt.savefig('rand-rayleigh.pdf', bbox_inches='tight', dpi=1000)
```
