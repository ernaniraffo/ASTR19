# Unit 2 Code Demo

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(5 * x)
```

```python
plt.plot(x, y)
plt.ylim((-1.1, 1.1))

# label axes
plt.xlabel('x')
plt.ylabel('y')

plt.savefig('demo.png', bbox_inches='tight', dpi=400, facecolor='lightgray')
```
